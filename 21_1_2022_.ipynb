```python
import pandas as pd

df=pd.read_csv('/content/iris.csv')
print(df[10:20])
print(df[["sepal_length","sepal_width","petal_length","petal_width"]])##retriving as dataframe
```

```python
#multivariate analysis
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,alpha=0.5,diagonal='kde')#kernal density estimation
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df,alpha=0.5,diagonal='hist')
```

```python
pd.plotting.scatter_matrix(df)#normal scatter matrix plot,plots are like histograms especially diagonal entries
```

Bar chart
bar function

```python
import matplotlib.pyplot as plt
langu=['Java','Python','PHP','JavaScript','C#','C++']
pop=[22.2,17.6,8,8,7.7,6.7]

plt.bar(langu,pop)
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.title("Popularity of Languages")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/iris.csv')
df.plot.bar()
plt.figure(figsize=(10,7))
#df.plot(kind='bar')
plt.show()
```

Bubble plot




```python
df=pd.read_csv('/content/iris.csv')
import matplotlib.pyplot as plt
plt.scatter('petal_length','petal_width',
           s='sepal_length',
           alpha=0.5,
           data=df)
plt.xlabel("petal_length",size=16)
plt.ylabel("petal_width",size=16)
plt.title("Bubble plot with Matplotlib",size=18)
```

```python
import seaborn as sns

iris=sns.load_dataset('iris')
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,height=4)#jointplot
```
